```python
# This is the Cache Function Jon wrote
import os
from requests import get
from urllib.parse import urlparse

def cache_data(src:str, dest:str) -> str:
    """Downloads and caches a remote file locally.
    
    The function sits between the 'read' step of a pandas or geopandas
    data frame and downloading the file from a remote location. The idea
    is that it will save it locally so that you don't need to remember to
    do so yourself. Subsequent re-reads of the file will return instantly
    rather than downloading the entire file for a second or n-th itme.
    
    Parameters
    ----------
    src : str
        The remote *source* for the file, any valid URL should work.
    dest : str
        The *destination* location to save the downloaded file.
        
    Returns
    -------
    str
        A string representing the local location of the file.
    """
    
    url = urlparse(src) # We assume that this is some kind of valid URL 
    fn  = os.path.split(url.path)[-1] # Extract the filename
    dfn = os.path.join(dest,fn) #Â Destination filename
    
    # Check if dest+filename does *not* exist -- 
    # that would mean we have to download it!
    if not os.path.isfile(dfn) or os.path.getsize(dfn) < 1:
        
        print(f"{dfn} not found, downloading!")

        # Convert the path back into a list (without)
        # the filename -- we need to check that directories
        # exist first.
        path = os.path.split(dest)
        
        # Create any missing directories in dest(ination) path
        # -- os.path.join is the reverse of split (as you saw above)
        # but it doesn't work with lists... so I had to google how
        # to use the 'splat' operator! os.makedirs creates missing
        # directories in a path automatically.
        if len(path) >= 1 and path[0] != '':
            os.makedirs(os.path.join(*path), exist_ok=True)
            
        # Download and write the file
        with open(dfn, "wb") as file:
            response = get(src)
            file.write(response.content)
            
        print('Done downloading...')

    else:
        print(f"Found {dfn} locally!")

    return dfn
```

### Adding Geography

```python
# some shapefiles
spath = 'https://github.com/jreades/fsds/blob/master/data/src/' # source path
ddir  = os.path.join('data','geo') # destination directory
water = gpd.read_file( cache_data(spath+'Water.gpkg?raw=true', ddir) )
boros = gpd.read_file( cache_data(spath+'Boroughs.gpkg?raw=true', ddir) )
green = gpd.read_file( cache_data(spath+'Greenspace.gpkg?raw=true', ddir) )
```

```python
boros.head(5)
```

```python
evict2019_geom = pd.merge(evictions2019, boros, how="right", left_on='Area', right_on='NAME')
evict2022_geom = pd.merge(evictions2022, boros, how="right", left_on='Area', right_on='NAME')
```

```python
evict_sums2019 = evict2019_geom['Repossessions']
evict_sums2019.plot.hist(bins=100)
```

```python
evict_sums2022 = evict2022_geom['Repossessions']
evict_sums2022.plot.hist(bins=100)
```

```python
evict2019_geom.info(verbose=True)
```

```python
evict2019_geom = gpd.GeoDataFrame(data=evict2019_geom, geometry='geometry', crs='epsg:27700')
gpd.GeoSeries(evict2019_geom.geometry)
```

```python
evict2022_geom = gpd.GeoDataFrame(data=evict2022_geom, geometry='geometry', crs='epsg:27700')
gpd.GeoSeries(evict2022_geom.geometry)
```

```python
evict2019_geom = evict2019_geom.to_crs('epsg:27700')
evict2022_geom = evict2022_geom.to_crs('epsg:27700')
```

```python
evict2019_geom['area'] = evict2019_geom.area
```

```python
evict2022_geom['area'] = evict2022_geom.area
```

```python
evict2019_geom['density'] =  (evict2019_geom['Repossessions'] / evict2019_geom['area'])*1e6
#1e6 converts to km sq not m sq
```

```python
evict2022_geom['density'] =  (evict2022_geom['Repossessions'] / evict2022_geom['area'])*1e6
#1e6 converts to km sq not m sq
```

```python
#Kernel Density based on Evictions 2019
from mpl_toolkits.axes_grid1 import make_axes_locatable
def setColorbar(ax, colormap, vmin, vmax):
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    sm = plt.cm.ScalarMappable(cmap=colormap, norm=plt.Normalize(vmin=vmin, vmax=vmax))
    # fake up the array of the scalar mappable. Urgh...
    sm._A = []
    return plt.colorbar(sm, cax=cax)

vmin=0
vmax=15
cmap='viridis'
f,ax=plt.subplots(figsize=(10,10))
evict2019_geom.plot(column='density', alpha=0.5,ax=ax, cmap=cmap, vmin=vmin, vmax=vmax)
setColorbar(ax, cmap, vmin=vmin, vmax=vmax)
```
